Q1. What is a contingency matrix, and how is it used?

A contingency matrix (also called a contingency table or cross-tabulation) shows how two categorical variables relate to each other.

In classification (or clustering evaluation), it usually means:

rows → true labels

columns → predicted labels

Each cell contains the number of samples that fall into that (true, predicted) combination.

How it is used

It is used to:

see how predictions are distributed across true classes

compute evaluation metrics such as:

accuracy

precision

recall

F1-score

A confusion matrix is actually a special case of a contingency matrix.

✅ Q2. Pair confusion matrix vs regular confusion matrix
Regular confusion matrix

Counts classification results at the individual sample level:

how many samples are correctly or incorrectly classified.

Pair confusion matrix

Instead of samples, it considers pairs of samples.

For clustering evaluation:

It checks whether a pair of points:

is in the same true class

and is in the same predicted cluster

It counts:

pairs correctly grouped together

pairs correctly separated

and the corresponding errors

Why is it useful?

A pair confusion matrix is useful when:

you evaluate clustering results

and you care about relationships between points, not just labels

It is the basis for metrics such as:

Rand Index

Adjusted Rand Index

✅ Q3. What is an extrinsic measure in NLP?

An extrinsic measure evaluates a language model or representation by using it in a real downstream task.

In simple words:

You test the model by seeing how well it helps another task.

Typical examples

use word embeddings in a sentiment classifier

use a language model in a named-entity recognition system

use a text representation for document classification

The performance on that task (accuracy, F1, etc.) is the extrinsic evaluation.

✅ Q4. What is an intrinsic measure and how is it different?

An intrinsic measure evaluates a model or representation directly, without using another task.

Key difference
Intrinsic measure	Extrinsic measure
Evaluates internal quality	Evaluates usefulness in a task
No downstream model needed	Requires a downstream task
Faster and simpler	More realistic but slower

Example:

evaluating word embeddings using word similarity scores → intrinsic

using those embeddings in a classifier → extrinsic

✅ Q5. Purpose of a confusion matrix in machine learning

The confusion matrix shows:

how many predictions fall into each true–predicted class combination

It helps you identify:

which classes are being confused

which classes are predicted well

which types of errors are common

How it shows strengths and weaknesses

From the matrix you can see:

high true positives → model is strong for that class

high false negatives → model misses that class

high false positives → model over-predicts that class

This is much more informative than only using accuracy.

✅ Q6. Common intrinsic measures for unsupervised learning

These do not require true labels.

Common intrinsic clustering measures

Silhouette Coefficient

Davies–Bouldin Index

Calinski–Harabasz Index

How to interpret them

Silhouette

higher → better separated and compact clusters

Davies–Bouldin

lower → better clustering

Calinski–Harabasz

higher → better clustering

They measure:

how tight clusters are

and how well clusters are separated

✅ Q7. Limitations of accuracy as a sole metric

Accuracy can be misleading, especially when data is imbalanced.

Main limitations
1. Class imbalance problem

Example:

95% samples belong to class A

model predicts everything as A

Accuracy = 95%
But the model is useless for the minority class.

2. Does not show error types

Accuracy does not tell you:

whether errors are false positives or false negatives

which class is being confused

3. Not suitable when costs of errors differ

In many real problems:

false negative and false positive have very different impact
(e.g., fraud detection, medical diagnosis)

How to address these limitations

Use additional metrics:

precision

recall

F1-score

ROC-AUC

class-wise confusion matrix

and always inspect the confusion matrix instead of relying only on accuracy.
